```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
# from webdriver_manager.chrome import ChromeDriverManager

# M·ªü trang web c·∫ßn qu√©t
def extract_comments(link):
    driver = webdriver.Chrome()
    driver.get(link)  # Thay URL_c·ªßa_trang_web b·∫±ng URL c·ªßa trang b·∫°n mu·ªën qu√©t

    # T√¨m t·∫•t c·∫£ c√°c ph·∫ßn t·ª≠ v·ªõi c√°c class b·∫°n ch·ªâ ƒë·ªãnh
    elements = driver.find_elements(By.CSS_SELECTOR, "div.xrvj5dj.xd0jker.x1evr45z")

    # L·∫∑p qua t·∫•t c·∫£ c√°c ph·∫ßn t·ª≠ v√† tr√≠ch xu·∫•t th√¥ng tin
    comments = []
    for element in elements:
        try:
            # Tr√≠ch xu·∫•t t√™n ng∆∞·ªùi d√πng (aoi.usagi)
            username = element.find_element(By.CSS_SELECTOR, "span.x1lliihq").text
            
            # Tr√≠ch xu·∫•t m√¥ t·∫£
            description = element.find_element(By.CSS_SELECTOR, "span.x1lliihq").text
            
            # Th√™m d·ªØ li·ªáu v√†o danh s√°ch comments d∆∞·ªõi d·∫°ng dictionary
            comments.append({"username": username, "description": description})

        except Exception as e:
            print(f"L·ªói khi tr√≠ch xu·∫•t t·ª´ ph·∫ßn t·ª≠: {e}")
    return comments

# V√≠ d·ª• g·ªçi h√†m
link = "https://www.threads.net/@vinhdidau/post/C-XqaynvPQP"  # Thay th·∫ø b·∫±ng link c·ª• th·ªÉ
comments_data = extract_comments(link)

# In ra k·∫øt qu·∫£
for comment in comments_data:
    print(comment)

# ƒê√≥ng tr√¨nh duy·ªát
driver.quit()
```
